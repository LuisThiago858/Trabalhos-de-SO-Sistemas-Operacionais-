```python
import numpy as np
import matplotlib.pyplot as plt

# Tamanhos mdios dos processos (em unidades de tempo)
tamanhos_processos = np.arange(10, 110, 10)  # De 10 a 100 unidades de tempo

# Simulao do tempo de resposta no SJF para diferentes tamanhos mdios de processo
# Supondo que quanto maior o tamanho mdio dos processos, maior ser o tempo de resposta
tempo_de_resposta = tamanhos_processos * 1.5  # Exemplo de uma funo linear simples para o grfico

# Criando o grfico 1
plt.figure(figsize=(8, 6))
plt.plot(tamanhos_processos, tempo_de_resposta, marker='o', color='r', label='Tempo de Resposta')
plt.title('Grfico 1: Relao entre Tamanho Mdio dos Processos e Tempo de Resposta (SJF)')
plt.xlabel('Tamanho Mdio dos Processos')
plt.ylabel('Tempo de Resposta (unidades de tempo)')
plt.grid(True)
plt.legend()

# Exibindo o grfico
plt.show()
```
