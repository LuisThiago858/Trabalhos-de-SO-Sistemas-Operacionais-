```python
import numpy as np
import matplotlib.pyplot as plt

# Tamanhos médios dos processos (em unidades de tempo)
tamanhos_processos = np.arange(10, 110, 10)  # De 10 a 100 unidades de tempo

# Simulação do tempo de resposta no SJF para diferentes tamanhos médios de processo
# Supondo que quanto maior o tamanho médio dos processos, maior será o tempo de resposta
tempo_de_resposta = tamanhos_processos * 1.5  # Exemplo de uma função linear simples para o gráfico

# Criando o gráfico 1
plt.figure(figsize=(8, 6))
plt.plot(tamanhos_processos, tempo_de_resposta, marker='o', color='r', label='Tempo de Resposta')
plt.title('Gráfico 1: Relação entre Tamanho Médio dos Processos e Tempo de Resposta (SJF)')
plt.xlabel('Tamanho Médio dos Processos')
plt.ylabel('Tempo de Resposta (unidades de tempo)')
plt.grid(True)
plt.legend()

# Exibindo o gráfico
plt.show()
```
